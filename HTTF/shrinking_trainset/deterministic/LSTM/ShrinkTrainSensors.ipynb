```python
import os
import pandas as pd

def shrink_train_set(train: pd.DataFrame,
                     valid: pd.DataFrame,
                     test: pd.DataFrame,
                     N: int,
                     seq_len: int = 5485):
    """
    Remove the last N*seq_len rows from train,
    leave valid and test unchanged.
    Saves each split as CSVs in "shrink_train_{N}" and prints lengths before/after.
    Returns new (train, valid, test) DataFrames.
    """
    # Print lengths before
    print(f"\n--- N = {N} ---")
    print("Before lengths:")
    print(f"  train: {len(train)}")
    print(f"  valid: {len(valid)}")
    print(f"  test : {len(test)}")

    step = N * seq_len
    train_new = train.iloc[:-step].reset_index(drop=True)
    
    # lengths after
    print("After lengths:")
    print(f"  train: {len(train_new)}")
    print(f"  valid: {len(valid)} (unchanged)")
    print(f"  test : {len(test)}  (unchanged)")

    # Create output directory and save
    output_dir = f"shrink_train_{N}"
    os.makedirs(output_dir, exist_ok=True)
    train_new.to_csv(os.path.join(output_dir, 'train.csv'), index=False)
    valid.to_csv(os.path.join(output_dir, 'valid.csv'), index=False)
    test.to_csv(os.path.join(output_dir, 'test.csv'), index=False)
    print(f"Saved splits to '{output_dir}/'")

    return train_new, valid, test

# Load and preprocess once
train_orig = pd.read_csv('TF_TS_train.csv')
valid_orig = pd.read_csv('TF_TS_valid.csv')
test_orig  = pd.read_csv('TF_TS_test.csv')

for df in (train_orig, valid_orig, test_orig):
    df['TS'] = df['TS'].interpolate()

# Loop over N = 1 to 62 sensors
for N in range(1, 63):
    shrink_train_set(train_orig, valid_orig, test_orig, N)
```
